```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000)
    return tracks
```

**quite a few duplicates and rows with none as values, will drop these**

```python
# drop dupes measured by artist and title

ultimate_df = ultimate_df.drop_duplicates(
      subset = ['title', 'artist'],
      keep = 'last').reset_index(drop = True)

# drop remaining 'none' rows
ultimate_df = ultimate_df.mask(ultimate_df.eq('none')).dropna()
```

```python
sp.audio_features(ultimate_df.at[0, 'uri'])
```

```python
# for loop to grab audio features per uri and drop them into a list

feature_list = []
for i in ultimate_df.index:
    feature_list.append(sp.audio_features(ultimate_df.at[i, 'uri']))
```

```python
# making a copy for safekeeping incase I mess things up
feature_list_copy=feature_list.copy()
```

```python
# feature_list
```

```python
import pandas as pd
pd.set_option("display.max_rows", 0)
```

```python
# turning into df

features = pd.DataFrame(feature_list, ignore_index = True)
features.columns = ['features']
features
```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
nested_columns = ['features']
```

```python
# flattening

features_flat = flatten(features, nested_columns)
features_flat
```

```python
features_flat.info()
```

```python
spotify_df = pd.concat([ultimate_df.reset_index(drop=True), features_flat.reset_index(drop=True)], axis =1)
```
